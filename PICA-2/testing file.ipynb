```python
import requests
import matplotlib.pylab as plt
import cv2
import numpy as np

url = 'http://127.0.0.1:8090/super?alpha=house&beta=butterfly&color=hot&name=Nico&noise_dim=100&num_examples=1'
final_output = requests.post(url)
output = final_output.content

i = np.frombuffer(output,dtype=np.uint8)
im = cv2.imdecode(i,cv2.IMREAD_UNCHANGED)
plt.imshow(im)
plt.axis('off')

```
